# Problem Statement:
####  Our client is a large MNC and they have 9 broad verticals across the organization. One of the problem your client is facing is around identifying the right people for promotion (only for manager position and below) and prepare them in time. Currently the process, they are following is: 
#### 1. They first identify a set of employees based on recommendations/ past performance#### 
2.	Selected employees go through the separate training and evaluation program for each vertical. These programs are based on the required skill of each vertical#### 
3.	At the end of the program, based on various factors such as training performance, KPI completion (only employees with KPIs completed greater than 60% are considered) etc., employee gets promotion


##### For above mentioned process, the final promotions are only announced after the evaluation and this leads to delay in transition to their new roles. Hence, company needs your help in identifying the eligible candidates at a particular checkpoint so that they can expedite the entire promotion cycle.They have provided multiple attributes around Employee's past and current performance along with demographics. Now, the task is to predict whether a potential promote at checkpoint in the test set will be promoted or not after the evaluation process.

#### 1. They first identify a set of employees based on recommendations/ past performance
#####  In step 1, I shortlisted employees based on historical performance indicators such as ratings, KPI completion, and experience.



```python
Shortlisted = df[
                (df['previous_year_rating']>=3) &
                (df['length_of_service']>2) &
                (df['KPIs_met'] == 1)
]

print("Before Selected total employees",df.shape)
print()
print("Ofter selected a set of employees",Shortlisted.shape)
```

#### For the first stage, I shortlisted employees based on historical performance indicators. This included previous year ratings, KPI completion status as per business requirement, and minimum organizational experience to ensure adequate exposure

### 2. Training and evaluation program for each vertical

```python
# 2.1: training participation by vertical 
department_vs_tp = Shortlisted.groupby('department')['no_of_trainings'].mean().sort_values(ascending = False)
# 2.2.: training evaluation by vertical
department_vs_te = Shortlisted.groupby('department')['avg_training_score'].mean().sort_values(ascending = False)
```

### 2.3: Selecting employees who passed training & evaluation

```python
evalution_data = Shortlisted[(Shortlisted['no_of_trainings']>1) & (Shortlisted['avg_training_score']>=60)]
evalution_data.shape
```

```python
print("Employees who performed well in training",evalution_data.shape[0])
```

##### Since each vertical has different skill requirements, department-wise training participation and evaluation scores were analyzed. This helps assess how employees performed in their respective vertical training programs. Employees who completed multiple trainings and achieved a minimum evaluation score were considered to have successfully passed the training and evaluation phase.

## Final promotion eligibility

```python
final_candidates = evalution_data[
    evalution_data['KPIs_met'] == 1
]
final_candidates.shape
```

```python
evalution_data.rename(
    columns = {
        'KPIs_met >80%' : 'KPIs_met',
        'awards_won?'   : 'awards_won'
    },
    inplace = True
)
```

```python
# Strong training performance
final_candidates = final_candidates[
    final_candidates['avg_training_score'] >= 70
]
final_candidates.shape
```

```python
# Include additional factors
final_candidates = final_candidates[
    (final_candidates['awards_won'] == 1) |
    (final_candidates['previous_year_rating'] >= 4)
]
final_candidates.shape
```

```python
final_candidates['is_promoted'].value_counts()
```

```python
final_candidates['is_promoted'].value_counts(normalize=True) * 100
```

#### In the final stage, promotion eligibility was determined based on KPI completion, training performance, and additional performance indicators such as awards and previous year ratings. Only employees meeting the KPI threshold and demonstrating strong training outcomes were considered promotion-ready
